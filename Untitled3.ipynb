```python
#Import Libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ML Models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB

# DL Models (Using TensorFlow 2.19 integrated Keras)
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Embedding, Conv1D, GlobalMaxPooling1D,
                                     Dense, Dropout, Bidirectional, LSTM)
from tensorflow.keras.callbacks import EarlyStopping

print("TensorFlow version:", tf.__version__)
```

```python
#Load data
from google.colab import files

print("DDI_data.csv")
uploaded = files.upload()

df = pd.read_csv("DDI_data.csv")
print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head())
```

```python
# Restrict to top 20 interaction types (by frequency) to reduce the effect of class imbalance and improve model learning
# This ensures that each class has sufficient training samples and allows the models to learn meaningful patterns.

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

top_20_labels = df['interaction_type'].value_counts().head(20).index
df_top = df[df['interaction_type'].isin(top_20_labels)].reset_index(drop=True)

print("Total records (top 20):", df_top.shape[0])
print("Unique interaction types (top 20):", df_top['interaction_type'].nunique())
print(df_top['interaction_type'].value_counts().head())

# Encode labels
label_encoder = LabelEncoder()
df_top['label'] = label_encoder.fit_transform(df_top['interaction_type'])
num_classes = len(label_encoder.classes_)
print("Encoded classes:", num_classes)

X_text = df_top['description'].values
y = df_top['label'].values

# 70 / 15 / 15 stratified split
X_temp, X_test, y_temp, y_test = train_test_split(
    X_text, y,
    test_size=0.15,
    random_state=RANDOM_SEED,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.17647,  # 0.17647 * 0.85 ≈ 0.15 -> overall 70/15/15
    random_state=RANDOM_SEED,
    stratify=y_temp
)

print("Train size:", len(X_train))
print("Val size:", len(X_val))
print("Test size:", len(X_test))
print("Train label distribution:")
print(pd.Series(y_train).value_counts())
```

```python
#TF-IDF features for ML models
from sklearn.feature_extraction.text import TfidfVectorizer

print("TF-IDF features")
tfidf_vectorizer = TfidfVectorizer(
    max_features=20000,
    ngram_range=(1, 2),
    lowercase=True
)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_val_tfidf = tfidf_vectorizer.transform(X_val)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print("TF-IDF shapes:")
print("Train:", X_train_tfidf.shape)
print("Val  :", X_val_tfidf.shape)
print("Test :", X_test_tfidf.shape)
```

```python
# Evaluate Key metrics
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report
)

import matplotlib.pyplot as plt

def evaluate_and_print(name, y_true, y_pred, train_scores=None, show_report=False):
    acc = accuracy_score(y_true, y_pred)
    macro_precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    macro_recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
    macro_f1 = f1_score(y_true, y_pred, average='macro')
    weighted_f1 = f1_score(y_true, y_pred, average='weighted')

    print(f"\n=== {name} ===")
    print(f"Final Test - Accuracy: {acc:.4f}, Macro F1: {macro_f1:.4f}, Weighted F1: {weighted_f1:.4f}")

    if show_report:
        print("\nDetailed Classification Report:")
        print(classification_report(y_true, y_pred, digits=4))

    if train_scores is not None:
        plt.figure(figsize=(10, 4))
        plt.plot(train_scores['train'], label='Train F1')
        plt.plot(train_scores['val'], label='Val F1')
        plt.title(f'{name} - Train vs Validation F1 (Plateau reached)')
        plt.xlabel('Iteration/Epoch')
        plt.ylabel('Macro F1 Score')
        plt.legend()
        plt.grid(True)
        plt.show()

    return acc, macro_precision, macro_recall, macro_f1, weighted_f1
```

```python
# Training Machine Learning Models
# Logistic Regression (LR)
import warnings
warnings.filterwarnings("ignore")  # Hide convergence warnings

results = []

print("\n=== Logistic Regression - Training until Plateau ===")
best_val_f1 = 0
patience_counter = 0
patience = 15
max_epochs = 200
epoch_scores = {'train_f1': [], 'val_f1': [], 'train_acc': [], 'val_acc': []}

log_reg = LogisticRegression(
    max_iter=50,  # steps per epoch
    class_weight='balanced',
    warm_start=True,
    random_state=42
)

for epoch in range(max_epochs):
    log_reg.fit(X_train_tfidf, y_train)

    # Track all metrics
    train_pred = log_reg.predict(X_train_tfidf)
    val_pred = log_reg.predict(X_val_tfidf)

    epoch_scores['train_f1'].append(f1_score(y_train, train_pred, average='macro'))
    epoch_scores['val_f1'].append(f1_score(y_val, val_pred, average='macro'))
    epoch_scores['train_acc'].append(accuracy_score(y_train, train_pred))
    epoch_scores['val_acc'].append(accuracy_score(y_val, val_pred))

    # Metric monitoring
    if f1_score(y_val, val_pred, average='macro') > best_val_f1:
        best_val_f1 = f1_score(y_val, val_pred, average='macro')
        patience_counter = 0
    else:
        patience_counter += 1

    if patience_counter >= patience:
        break

# Graph visualisaztion
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epoch_scores['train_f1'], label='Train F1', alpha=0.7)
plt.plot(epoch_scores['val_f1'], label='Val F1', linewidth=3)
plt.title('F1 Convergence (Plateau Reached)')
plt.ylabel('Macro F1')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epoch_scores['train_acc'], label='Train Acc', alpha=0.7)
plt.plot(epoch_scores['val_acc'], label='Val Acc', linewidth=3)
plt.title('Accuracy Convergence')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Metrics Output
y_pred_lr = log_reg.predict(X_test_tfidf)
acc = accuracy_score(y_test, y_pred_lr)
macro_precision = precision_score(y_test, y_pred_lr, average='macro', zero_division=0)
macro_recall = recall_score(y_test, y_pred_lr, average='macro', zero_division=0)
macro_f1 = f1_score(y_test, y_pred_lr, average='macro')
weighted_f1 = f1_score(y_test, y_pred_lr, average='weighted')

print("\n" + "="*50)
print("LOGISTIC REGRESSION - FINAL RESULTS")
print("="*50)
print(f"Accuracy        : {acc:.4f}")
print(f"Macro Precision : {macro_precision:.4f}")
print(f"Macro Recall    : {macro_recall:.4f}")
print(f"Macro F1        : {macro_f1:.4f}")
print(f"Weighted F1     : {weighted_f1:.4f}")
print(f"Plateau epoch   : {len(epoch_scores['val_f1'])}")
print("="*50)

res_lr = (acc, macro_precision, macro_recall, macro_f1, weighted_f1)
results.append(("LogisticRegression",) + res_lr)
```

```python
# Support Vector Machine (SVM)
import warnings
warnings.filterwarnings("ignore")

print("\n=== SVM - Training until Plateau ===")
best_val_f1 = 0
patience_counter = 0
patience = 15
max_epochs = 200
svm_scores = {'train_f1': [], 'val_f1': [], 'train_acc': [], 'val_acc': []}

from sklearn.svm import LinearSVC
svm_clf = LinearSVC(class_weight='balanced', max_iter=50, random_state=42)

for iteration in range(max_epochs):
    svm_clf.fit(X_train_tfidf, y_train)

    train_pred = svm_clf.predict(X_train_tfidf)
    val_pred = svm_clf.predict(X_val_tfidf)

    svm_scores['train_f1'].append(f1_score(y_train, train_pred, average='macro'))
    svm_scores['val_f1'].append(f1_score(y_val, val_pred, average='macro'))
    svm_scores['train_acc'].append(accuracy_score(y_train, train_pred))
    svm_scores['val_acc'].append(accuracy_score(y_val, val_pred))

    if f1_score(y_val, val_pred, average='macro') > best_val_f1:
        best_val_f1 = f1_score(y_val, val_pred, average='macro')
        patience_counter = 0
    else:
        patience_counter += 1

    if patience_counter >= patience:
        break

# Graphs
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(svm_scores['train_f1'], label='Train F1')
plt.plot(svm_scores['val_f1'], label='Val F1', linewidth=3)
plt.title('SVM F1 Convergence')
plt.legend(); plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(svm_scores['train_acc'], label='Train Acc')
plt.plot(svm_scores['val_acc'], label='Val Acc', linewidth=3)
plt.title('SVM Accuracy Convergence')
plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()

# Final metrics
y_pred_svm = svm_clf.predict(X_test_tfidf)
acc = accuracy_score(y_test, y_pred_svm)
macro_precision = precision_score(y_test, y_pred_svm, average='macro', zero_division=0)
macro_recall = recall_score(y_test, y_pred_svm, average='macro', zero_division=0)
macro_f1 = f1_score(y_test, y_pred_svm, average='macro')
weighted_f1 = f1_score(y_test, y_pred_svm, average='weighted')

print("\n" + "="*50)
print("LINEAR SVM - FINAL RESULTS")
print("="*50)
print(f"Accuracy        : {acc:.4f}")
print(f"Macro Precision : {macro_precision:.4f}")
print(f"Macro Recall    : {macro_recall:.4f}")
print(f"Macro F1        : {macro_f1:.4f}")
print(f"Weighted F1     : {weighted_f1:.4f}")
print(f"Plateau iter    : {len(svm_scores['val_f1'])}")
print("="*50)

results.append(("LinearSVM",) + (acc, macro_precision, macro_recall, macro_f1, weighted_f1))
```

```python
# Random Forest
print("\nTraining Random Forest...")
rf_scores = {'train_f1': [], 'val_f1': [], 'train_acc': [], 'val_acc': []}
n_estimators_range = [50, 100, 150, 200, 250, 300, 350, 400]
best_val_f1 = 0
best_rf = None

from sklearn.ensemble import RandomForestClassifier
for n_est in n_estimators_range:
    rf_clf = RandomForestClassifier(n_estimators=n_est, max_depth=20, n_jobs=-1,
                                   class_weight='balanced_subsample', random_state=42)
    rf_clf.fit(X_train_tfidf, y_train)

    train_pred = rf_clf.predict(X_train_tfidf)
    val_pred = rf_clf.predict(X_val_tfidf)

    train_f1 = f1_score(y_train, train_pred, average='macro')
    val_f1 = f1_score(y_val, val_pred, average='macro')

    rf_scores['train_f1'].append(train_f1)
    rf_scores['val_f1'].append(val_f1)
    rf_scores['train_acc'].append(accuracy_score(y_train, train_pred))
    rf_scores['val_acc'].append(accuracy_score(y_val, val_pred))

    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        best_rf = rf_clf

# Graphs
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(rf_scores['train_f1'], 'o-', label='Train F1')
plt.plot(rf_scores['val_f1'], 'o-', label='Val F1', linewidth=3)
plt.title('RF F1 vs Trees')
plt.xlabel('Trees'); plt.ylabel('Macro F1'); plt.legend(); plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(rf_scores['train_acc'], 'o-', label='Train Acc')
plt.plot(rf_scores['val_acc'], 'o-', label='Val Acc', linewidth=3)
plt.title('RF Accuracy vs Trees')
plt.xlabel('Trees'); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()

# Final metrics using best model
y_pred_rf = best_rf.predict(X_test_tfidf)
acc = accuracy_score(y_test, y_pred_rf)
macro_precision = precision_score(y_test, y_pred_rf, average='macro', zero_division=0)
macro_recall = recall_score(y_test, y_pred_rf, average='macro', zero_division=0)
macro_f1 = f1_score(y_test, y_pred_rf, average='macro')
weighted_f1 = f1_score(y_test, y_pred_rf, average='weighted')

print("\n" + "="*50)
print("RANDOM FOREST - FINAL RESULTS")
print("="*50)
print(f"Accuracy        : {acc:.4f}")
print(f"Macro Precision : {macro_precision:.4f}")
print(f"Macro Recall    : {macro_recall:.4f}")
print(f"Macro F1        : {macro_f1:.4f}")
print(f"Weighted F1     : {weighted_f1:.4f}")
print(f"Best trees      : {n_estimators_range[len(rf_scores['val_f1'])-1]}")
print("="*50)

results.append(("RandomForest",) + (acc, macro_precision, macro_recall, macro_f1, weighted_f1))
```

```python
# Prepare sequences and embeddings for DL
# Transform raw text into numerical format that CNN/LSTM models can understand
print("\nTokenizing text for deep learning...")

MAX_VOCAB_SIZE = 20000      # for Tokenizer and TF-IDF
MAX_SEQUENCE_LENGTH = 50    # length of padded sequences

tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

word_index = tokenizer.word_index
vocab_size = min(MAX_VOCAB_SIZE, len(word_index) + 1)
print("Vocab size:", vocab_size)
```

```python
# Create neural network architectures for text classification.
def build_cnn_model():
    model = Sequential()
    model.add(Embedding(
        input_dim=vocab_size,
        output_dim=EMBEDDING_DIM,
        input_length=MAX_SEQUENCE_LENGTH,
        trainable=True
    ))
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

def build_bilstm_model():
    model = Sequential()
    model.add(Embedding(
        input_dim=vocab_size,
        output_dim=EMBEDDING_DIM,
        input_length=MAX_SEQUENCE_LENGTH,
        trainable=True
    ))
    model.add(Bidirectional(LSTM(128, return_sequences=False)))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model
```

```python
# Training Deep Learning Models
# Train Convolutional Neural Network (CNN)

import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping

print("\n=== CNN Training until Plateau ===")
EPOCHS = 30
BATCH_SIZE = 256            # Optimal batch size for training
EMBEDDING_DIM = 300         # GloVe dimension

cnn_model = build_cnn_model()
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)

history_cnn = cnn_model.fit(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop],
    verbose=1
)

# Graph Visualisation
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history_cnn.history['loss'], label='Train Loss', linewidth=2)
plt.plot(history_cnn.history['val_loss'], label='Val Loss', linewidth=3)
plt.title('CNN Loss Plateau'); plt.legend(); plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history_cnn.history['accuracy'], label='Train Acc', linewidth=2)
plt.plot(history_cnn.history['val_accuracy'], label='Val Acc', linewidth=3)
plt.title('CNN Accuracy Plateau'); plt.legend(); plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history_cnn.history['val_loss'], label='Val Loss', linewidth=3, color='red')
plt.axhline(y=min(history_cnn.history['val_loss']), color='green', linestyle='--', label='FLAT AREA')
plt.title('CNN Flat Area Proof'); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()

print(f"✓ CNN plateau reached at epoch {len(history_cnn.history['loss'])}")

# CNN - FINAL RESULTS
y_pred_cnn = np.argmax(cnn_model.predict(X_test_pad, verbose=0), axis=1)
cnn_acc = accuracy_score(y_test, y_pred_cnn)
cnn_f1 = f1_score(y_test, y_pred_cnn, average='macro')
print(f"CNN Test Accuracy: {cnn_acc:.4f}, Macro F1: {cnn_f1:.4f}")
```

```python
# Bidirectional LSTM (BiLSTM)

import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

print("\n=== BiLSTM Training until Plateau ===")
EPOCHS = 30
BATCH_SIZE = 256            # Optimal batch size for training
EMBEDDING_DIM = 300         # GloVe dimension

bilstm_model = build_bilstm_model()
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)

history_bilstm = bilstm_model.fit(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop],
    verbose=1
)

# Graph Visualisation
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history_bilstm.history['loss'], label='Train Loss', linewidth=2)
plt.plot(history_bilstm.history['val_loss'], label='Val Loss', linewidth=3)
plt.title('BiLSTM Loss Plateau'); plt.legend(); plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history_bilstm.history['accuracy'], label='Train Acc', linewidth=2)
plt.plot(history_bilstm.history['val_accuracy'], label='Val Acc', linewidth=3)
plt.title('BiLSTM Accuracy Plateau'); plt.legend(); plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history_bilstm.history['val_loss'], label='Val Loss', linewidth=3, color='red')
plt.axhline(y=min(history_bilstm.history['val_loss']), color='green', linestyle='--', label='FLAT AREA')
plt.title('BiLSTM Flat Area Proof'); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()

print(f"✓ BiLSTM plateau reached at epoch {len(history_bilstm.history['loss'])}")

# BiLSTM - FINAL RESULTS
y_pred_bilstm = np.argmax(bilstm_model.predict(X_test_pad, verbose=0), axis=1)
bilstm_acc = accuracy_score(y_test, y_pred_bilstm)
bilstm_f1 = f1_score(y_test, y_pred_bilstm, average='macro')
print(f"BiLSTM Test Accuracy: {bilstm_acc:.4f}, Macro F1: {bilstm_f1:.4f}")
```

```python
# ==================================================
# CONFUSION MATRIX + TP/FP/FN/TN TABLE - ALL MODELS
# ==================================================
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np

# Predictions
models = {
    'Linear SVM': y_pred_svm,
    'Random Forest': y_pred_rf,
    'Naive Bayes': y_pred_nb,
    'CNN': y_pred_cnn,
    'BiLSTM': y_pred_bilstm
}

# CONFUSION MATRIX GRAPHS
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.ravel()

for idx, (model_name, y_pred) in enumerate(models.items()):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx], cbar=False)
    axes[idx].set_title(f'{model_name}')
    axes[idx].set_xlabel('Predicted'); axes[idx].set_ylabel('Actual')

axes[-1].remove()
plt.suptitle('Confusion Matrices - All Models', fontsize=16)
plt.tight_layout()
plt.savefig('all_confusion_matrices.png', dpi=300)
plt.show()

# TP/FP/FN/TN TABLE
table_data = []
for model_name, y_pred in models.items():
    cm = confusion_matrix(y_test, y_pred)
    tp = np.trace(cm)
    fp = cm.sum(axis=0) - np.diagonal(cm)
    fn = cm.sum(axis=1) - np.diagonal(cm)
    tn = cm.sum() - (fp.sum() + fn.sum() + tp)

    table_data.append({
        'Model': model_name,
        'TP': tp, 'FP': fp.sum(), 'FN': fn.sum(), 'TN': tn,
        'Accuracy': f"{accuracy_score(y_test, y_pred):.1%}"
    })

cm_table = pd.DataFrame(table_data)
print("\nTABLE 4.4: TP/FP/FN/TN SUMMARY")
print("="*60)
print(cm_table.to_string(index=False, float_format='%.0f'))
print("="*60)

# Save table
cm_table.to_csv('confusion_table.csv', index=False)
print("✅ SAVED: all_confusion_matrices.png + confusion_table.csv")
```
